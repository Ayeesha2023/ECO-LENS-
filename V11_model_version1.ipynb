```python
# 1. Connect to Google Drive (if not already connected)
from google.colab import drive
drive.mount('/content/drive')

# 2. Install the latest version of Ultralytics (gets you YOLO11)
!pip install -U ultralytics

# 3. Import the library
from ultralytics import YOLO

# 4. Initialize the brand new YOLO11 Medium model (It will auto-download the weights)
# Notice we are starting fresh with 'yolo11m.pt', NOT loading our old YOLOv8 weights!
model = YOLO('yolo11m.pt')

# 5. Start Training with Advanced Settings for Small Objects & T4 GPU
results = model.train(
    data='/content/drive/MyDrive/ECO-LENS/Taco/data.yaml', # Check your exact spelling of Taco/TACO
    epochs=55,                                            # Full run
    imgsz=800,                                            # High-res mode for tiny objects (caps, cigarettes)
    device=0,                                             # Use the free T4 GPU
    project='/content/drive/MyDrive/ECO-LENS/runs',       # Save directly to your Drive
    name='ecolens_yolo11m_highres',                       # Name of the new folder
    patience=40,                                          # Safety net: waits 40 epochs before stopping if flatlined
    batch=8,                                              # LOWERED to 8: prevents the T4 GPU from running Out of Memory!
    optimizer='auto'                                      # Let YOLO pick the best learning optimizer natively
)
```
