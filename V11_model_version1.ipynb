```python
# Install Ultralytics
!pip install -U ultralytics

# Import library
from ultralytics import YOLO

# Load pretrained YOLO11 Medium model
model = YOLO("yolo11m.pt")

# Train the model
results = model.train(
    data="/content/drive/MyDrive/ECO-LENS/Taco/data.yaml",  # Change if needed

    epochs=100,
    patience=20,

    imgsz=640,
    batch=16,

    optimizer="auto",
    lr0=0.001,

    device=0,
    workers=2,
    amp=True,

    pretrained=True,

    project="/content/drive/MyDrive/ECO-LENS/runs",
    name="YOLO11m",

    save=True,
    verbose=True
)

# Validate
metrics = model.val()

print(metrics)
```
