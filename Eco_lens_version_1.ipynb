```python
# Load the small, pre-trained YOLOv8 model
model = YOLO('yolov8s.pt')
```

```python
# Start training the model
results = model.train(
    data='/content/drive/MyDrive/ECO-LENS/Taco/data.yaml', # Path to your edited data.yaml
    epochs=50,                                            # Number of times the model sees the entire dataset
    imgsz=640,                                            # Standard image resizing size for YOLO
    device=0,                                             # Tells it to use the T4 GPU
    project='/content/drive/MyDrive/ECO-LENS/runs',       # FOR SAFETY: Saves checkpoints directly to your Drive!
    name='waste_classifier'                               # Name of the training session folder
)
```
